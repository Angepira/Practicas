```python
import pandas as pd

dataset=pd.read_csv('salario.csv')
dataset
```

```python
X=dataset.iloc[:,1:2].values
X
y=dataset.iloc[:,2].values
y
```

```python
from sklearn.tree import DecisionTreeRegressor

salario=DecisionTreeRegressor(random_state=0)
salario.fit(X, y)
```

```python
import numpy as np
y_pred=salario.predict(np.array(15).reshape(-1,1))
y_pred
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

X_new=np.arange(min(X),max(X),0.1)[:,np.newaxis]
X_new
y_pred=salario.predict(X_new)
plt.figure(figsize=(10,6))
plt.scatter(X,y, edgecolors="black", c="orange", label="Datos")
plt.plot(X_new,y_pred, color="cornflowerblue",label="Prediccion")
plt.legend()
plt.show
```

```python
presicion=salario.score(X,y)
presicion
```
