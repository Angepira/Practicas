```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('csv/ingreso.csv')
data
```

```python
X = data['horas'].values.reshape(-1, 1)
X
```

```python
y = data['ingreso'].values
y
```

```python
model = LinearRegression()
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
result = cross_val_score(model, X, y, cv=kf, scoring="r2")
```

```python
print(f"Cross-validation scores: {result}")
print(f"Promedio de R2: {result.mean()}")
```
