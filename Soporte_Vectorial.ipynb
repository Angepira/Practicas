```python
from sklearn.datasets import load_breast_cancer

dataset= load_breast_cancer()
dataset

X= dataset.data
y= dataset.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3, random_state=0)
```

```python
from sklearn.svm import SVC

modelo= SVC(kernel='linear')
modelo.fit(X_train, y_train)
```

```python
y_predic=modelo.predict(X_test)
y_predic
```

```python
from sklearn.metrics import confusion_matrix

matriz= confusion_matrix(y_test,y_predic)
matriz
```

```python
from sklearn.metrics import precision_score

presicion=precision_score(y_test,y_predic)
presicion
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

presicion1, recuperacion, _=precision_recall_curve(y_test,y_predic)
presicion1

plt.plot(recuperacion, presicion1)
plt.xlabel('Recuperacion')
plt.ylabel('Presicion')
plt.show()
```
