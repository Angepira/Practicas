```python
from sklearn.datasets import fetch_openml
import pandas as pd

boston=fetch_openml(name="boston", as_frame=True)
boston.data
df_boston= pd.DataFrame(boston.data, columns=boston.feature_names)
df_boston
```

```python
boston.target_names
```

```python
import numpy as np

#Variable Independiente
X_adr=df_boston['RM'].values.reshape(-1,1)
X_adr

#Variable dependiente
y_adr=boston.target
y_adr=y_adr.values
X_adr
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X_adr,y_adr, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeRegressor

adr=DecisionTreeRegressor(max_depth=5)
```

```python
#Entrenamiento
adr.fit(X_train,y_train)
```

```python
Y_pred=adr.predict([[4],])
Y_pred
```

```python
presicion=adr.score(X_train,y_train)
presicion
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

X_new=np.arange(min(X_test),max(X_test),0.2)[:,np.newaxis]
X_new
y_pred=adr.predict(X_new)
plt.figure(figsize=(10,6))
plt.scatter(X_test,y_test, edgecolors="black", c="orange", label="Datos")
plt.plot(X_new,y_pred, color="cornflowerblue",label="Prediccion")
plt.legend()
plt.show
```
