```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns= iris.feature_names)
iris_df
```

```python
iris_df.isnull().sum()
```

```python
iris_df = iris_df.drop_duplicates()
iris_df
```

```python
from scipy import stats
import numpy as np
```

```python
iris_df = iris_df[(np.abs(stats.zscore(iris_df))<3).all(axis=1)] #elimina outlines
```

```python
from  sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

iris_escala = scaler.fit_transform(iris_df.iloc[:,:-1])

iris_escala
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters =3)
kmeans.fit_transform(iris_escala)
kmeans
```
