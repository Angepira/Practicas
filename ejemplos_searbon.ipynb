```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
df_sea=sns.load_dataset('flights')
df_sea
```

```python
df_sea.info()
```

```python
sns.scatterplot(data=df_sea, x='year', y='passengers', color='orange')
plt.title('Años vs Pasajeros')
plt.xlabel('Años')
plt.ylabel('Pasajeros')
plt.show()
```

```python
sns.scatterplot(data=df_sea, x='year', y='passengers', palette=sns.color_palette('husl',n_colors=12), hue='month')
plt.title('Años vs Pasajeros')
plt.xlabel('Años')
plt.ylabel('Pasajeros')
plt.show()
```

```python
df_sea2= sns.load_dataset('titanic')
df_sea2
```

```python
sns.barplot(data=df_sea2,x='class', y='fare', palette='dark', errorbar=None, order=['Third','First','Second'])
plt.title('Pagado por Clase')
plt.xlabel('Clase')
plt.ylabel('Pagado')
plt.show()
```

```python
sns.barplot(data=df_sea2,x='class', y='fare', palette='dark', errorbar=None, order=['Third','First','Second'])
plt.title('Pagado por Clase')
plt.xlabel('Clase')
plt.ylabel('Pagado')
plt.show()
```

```python
grafrico=sns.barplot(data=df_sea2,x='fare', y='class', palette='pastel', orient='h',errorbar=None, hue='sex')
plt.title('Pagado por Clase')
plt.xlabel('Pagado')
plt.ylabel('Tipo de Clase')
plt.show()
```

```python
sns.histplot(data=df_sea2, x='age')
plt.title('Edad de Psajeros')
plt.xlabel('Edad')
plt.show()
```

```python
sns.kdeplot(data=df_sea2, x='age', palette='flare', hue='sex')
plt.title('Edad de Psajeros')
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.show()
```

```python
sns.kdeplot(data=df_sea2, x='age', palette='Set2', hue='sex', multiple='stack')
plt.title('Edad de Psajeros')
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.show()
```

```python
sns.displot(data=df_sea2, x='age', palette='Set2', kde=True)
plt.title('Edad de Psajeros')
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.show()
```

```python
sns.displot(data=df_sea2, x='age', palette='dark', kind='kde', hue='class', fill=True)
plt.title('Edad de Psajeros')
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.show()
```

```python
sns.boxplot(data=df_sea2, x='class', y='age', palette='Spectral', hue='sex')
```

```python
sns.violinplot(data=df_sea2, x='class', y='age', palette='Spectral', hue='sex')
```
