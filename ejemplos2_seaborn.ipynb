```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_ping=sns.load_dataset('penguins')
df_ping
```

```python
sns.relplot(data=df_ping, x='bill_length_mm', y='flipper_length_mm',height=5, hue='body_mass_g')
```

```python
sns.relplot(data=df_ping, x='bill_length_mm', y='flipper_length_mm',height=5, hue='body_mass_g', s=100)
```

```python
sns.relplot(data=df_ping, x='bill_length_mm', y='flipper_length_mm',height=5, hue='species', s=100)
```

```python
sns.relplot(data=df_ping, x='bill_length_mm', y='flipper_length_mm',height=5, hue='species', row='sex', 
            size='body_mass_g', style='sex', sizes=(100,500), col='island')
```

```python
sns.relplot(data=df_ping, x='species', y='body_mass_g', kind='line', hue='sex')
plt.show()
sns.relplot(data=df_ping, x='flipper_length_mm', y='body_mass_g', kind='line', hue='species', style='species')
plt.show()
```

```python
df_ejercicio=sns.load_dataset('exercise')
df_ejercicio
```

```python
sns.set_theme(style='ticks', palette='tab10')
sns.catplot(data=df_ejercicio,x='kind',y='pulse', hue='kind', s=50, col='diet') #strip
plt.show()
```

```python
sns.set_theme(style='ticks', palette='tab10')
sns.catplot(data=df_ejercicio,x='kind',y='pulse', hue='kind', s=50, col='diet', kind='swarm') #strip
plt.show()
```

```python
df_mexico=pd.read_csv('Natalidad_Mexico_limpio.csv')
df_mexico=df_mexico.pivot(index='estado', columns='year',values='Total')
df_mexico
plt.figure(figsize=(15,15))
sns.heatmap(df_mexico, linewidths=0.5, annot=True, fmt='d')
```

```python
df_ping2=sns.load_dataset('penguins')
df_ping2=df_ping2.groupby(['island', 'species'], as_index=False).size()
df_ping2=df_ping2.pivot(index='island',columns='species',values='size')
df_ping2=df_ping2.fillna(0)
sns.heatmap(data=df_ping2, annot=True, fmt='.1f', linewidths=1, center=200)
df_ping2
plt.show()
```
