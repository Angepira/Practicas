```python
import pandas as pd

personas=pd.read_csv('ingresos.csv')
personas
```

```python
from sklearn.ensemble import RandomForestClassifier

bosque=RandomForestClassifier()
bosque.get_params()
```

```python
from sklearn.model_selection import GridSearchCV

parametros={'criterion':('gini','entropy'),
            'n_estimators':(10,20,30),
            'max_samples':(1/3,2/3),
            }
```

```python
rejilla= GridSearchCV(bosque,parametros,scoring='f1')

rejilla.fit(personas[personas.columns[:-1]].values,personas['ingreso'].values)
```

```python
rejilla.cv_results_['params']
```

```python
print(rejilla.cv_results_["rank_test_score"])
print(rejilla.cv_results_["mean_test_score"])
print(rejilla.best_score_)
print(rejilla.best_params_)
```

```python
rejilla.predict([[55, 17, 1, 0,40]])
mejor=rejilla.best_estimator_
mejor.predict([[55, 17, 1, 0,40]])
```

```python
#Tuning Aleatorio

from sklearn.model_selection import RandomizedSearchCV

parametros={'criterion':('gini','entropy'),
            'n_estimators':(10,20,30),
            'max_samples':(1/3,2/3),
            }
```

```python
rejilla_Aleatoria= RandomizedSearchCV(bosque,parametros,scoring='accuracy',n_iter=3)

rejilla_Aleatoria.fit(personas[personas.columns[:-1]].values,personas['ingreso'].values)
```

```python
print(rejilla_Aleatoria.best_score_)
print(rejilla_Aleatoria.best_params_)
```

```python
mejor_A=rejilla_Aleatoria.best_estimator_
mejor_A.predict([[55, 17, 1, 0,40]])
```
