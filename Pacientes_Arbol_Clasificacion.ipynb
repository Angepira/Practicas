```python
import pandas as pd

pacientes= pd.read_csv('Pacientes2.csv')
pacientes
```

```python
pacientes=pacientes.set_index('NOEXPED')
pacientes
```

```python
pacientes.info()
```

```python
X=pacientes.iloc[:,1:11]
X
```

```python
y=pacientes.iloc[:,0]
y
```

```python
from sklearn.model_selection import train_test_split

datos_train, datos_test, clase_train, clase_test=train_test_split(X,y,test_size=0.2)
datos_train
```

```python
clase_test
```

```python
from sklearn import tree

arbol_dec=tree.DecisionTreeClassifier(max_depth=5)
arbol_dec.fit(datos_train,clase_train)
```

```python
presicion= arbol_dec.score(datos_test,clase_test)
presicion
```

```python
import matplotlib.pyplot as plt

print(tree.export_text(arbol_dec,feature_names=list(X.columns.values)))
plt.figure(figsize=(12,8))
tree.plot_tree(arbol_dec, feature_names=list(X.columns.values), class_names=list(y.values), filled=True)
plt.show()
```

```python
Y_pred= arbol_dec.predict(datos_test)
Y_pred
```

```python
from sklearn.metrics import confusion_matrix

matriz=confusion_matrix(clase_test,Y_pred)
print(matriz)
```
